# Fandom.com

```python
import pandas as pd
import re
import string
import nltk
from nltk.corpus import stopwords
```

```python
file_path = r'C:\Users\josie\Downloads\FANDOM_DATA.csv'
```

```python
df = pd.read_csv(file_path)
```

```python
df
```

```python
# Define a function to lowercase text
def lowercase_text(text):
    if pd.isna(text):
        return text
    return text.lower()

# Define a function to remove punctuation
def remove_punctuation(text):
    if pd.isna(text):
        return text
    return text.translate(str.maketrans('', '', string.punctuation))

# Define a function to remove stopwords
def remove_stopwords(text):
    if pd.isna(text):
        return text
    stop_words = set(stopwords.words('english'))
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    return ' '.join(filtered_words)

# List of columns to clean
columns_to_clean = ['Synopsis', 'Biography', 'Physical appearance', 'Personality', 'Relationships']

# Apply the cleaning functions to each specified column
for column in columns_to_clean:
    df[column] = df[column].apply(lowercase_text)
    df[column] = df[column].apply(remove_punctuation)
    df[column] = df[column].apply(remove_stopwords)

# Display the first few rows of the cleaned DataFrame
print(df.head())
```

```python
# Save the cleaned DataFrame to a new text file
cleaned_file_path = r'C:\Users\josie\Downloads\FANDOM_DATA_CLEANED.txt'
df.to_csv(cleaned_file_path, index=False, sep='\t')

print(f"Cleaned data saved to {cleaned_file_path}")
```

```python
# Filter DataFrame based on gender
female_biographies = df[df['Gender'] == 'Female']['Biography']
male_biographies = df[df['Gender'] == 'Male']['Biography']

# Save the cleaned data to separate text files
female_biographies_file_path = r'C:\Users\josie\Downloads\FEMALE_Biographies_CLEANED.txt'
male_biographies_file_path = r'C:\Users\josie\Downloads\MALE_Biographies_CLEANED.txt'

female_biographies.to_csv(female_biographies_file_path, index=False, header=False, sep='\t')
male_biographies.to_csv(male_biographies_file_path, index=False, header=False, sep='\t')

print(f"Cleaned female biographies saved to {female_biographies_file_path}")
print(f"Cleaned male biographies saved to {male_biographies_file_path}")
```

```python
gender_columns = ['Physical appearance', 'Personality', 'Relationships']

for column in gender_columns:
    # Female
    female_data = df[df['Gender'] == 'Female'][column]
    female_file_path = f'C:\\Users\\josie\\Downloads\\FEMALE_{column.replace(" ", "_").upper()}_CLEANED.txt'
    female_data.to_csv(female_file_path, index=False, header=False, sep='\t')
    
    # Male
    male_data = df[df['Gender'] == 'Male'][column]
    male_file_path = f'C:\\Users\\josie\\Downloads\\MALE_{column.replace(" ", "_").upper()}_CLEANED.txt'
    male_data.to_csv(male_file_path, index=False, header=False, sep='\t')

    print(f"Cleaned female {column} data saved to {female_file_path}")
    print(f"Cleaned male {column} data saved to {male_file_path}")
```

```python
media_types = ['Book', 'Movie', 'Video Game', 'TV Show', 'Book to TV Show', 'Video Game to TV Show', 'Book to Movie', 'All']

# Save filtered data to separate text files based on media type
for media_type in media_types:
    filtered_data = df[df['Media Type'] == media_type]
    
    # Save the entire filtered DataFrame to a text file
    file_path = f'C:\\Users\\josie\\Downloads\\{media_type.replace(" ", "_").upper()}_CLEANED.txt'
    filtered_data.to_csv(file_path, index=False, sep='\t')
    
    print(f"Cleaned data for {media_type} saved to {file_path}")
```
